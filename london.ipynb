```python
import requests
import json
import pandas as pd 
```

```python
base_url = "https://data.police.uk/api/stops-street?"
location = "lat=51.5072&lng=-0.1276&date="
dates = ["2021-01","2021-02","2021-03","2021-04","2021-05"]
```

```python
jan_may = []
for date in dates:
    query_url = base_url + location + date
    response = requests.get(query_url).json()
    jan_may.append(response)
```

```python
london_data = pd.DataFrame.from_dict(jan_may[0])
london_data = london_data.append(pd.DataFrame.from_dict(jan_may[1]))
london_data = london_data.append(pd.DataFrame.from_dict(jan_may[2]))
london_data = london_data.append(pd.DataFrame.from_dict(jan_may[3]))
london_data = london_data.append(pd.DataFrame.from_dict(jan_may[4]))
london_data
```

```python
london_info = london_data[["age_range", "self_defined_ethnicity", "gender", "officer_defined_ethnicity", "object_of_search", "location"]]
london_info
```

```python
london_info['country'] = 'london'
```

```python
london_info.set_index('country', inplace=True)
london_info
```

```python
age_range = london_info["age_range"]
age_range.groupby(age_range).count()
```

```python
outcome = london_info["object_of_search"]
outcome.groupby(outcome).count()
```

```python
self_defined_ethnicity = london_info["self_defined_ethnicity"]
self_defined_ethnicity.groupby(self_defined_ethnicity).count()
```

```python
officer_defined_ethnicity = london_info["officer_defined_ethnicity"]
officer_defined_ethnicity.groupby(officer_defined_ethnicity).count()
```

```python
gender = london_info["gender"]
gender.groupby(gender).count()
```
