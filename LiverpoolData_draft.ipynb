```python
# parameters dictionary
params = {
    "lat": 53.400002,
    "lng": -2.983333,
    "date": "2021-01",
}
```

```python
base_url = "https://data.police.uk/api/stops-street?"
jan_response = requests.get(base_url, params=params).json()
```

```python
jan_response
```

```python
liverpool_1 = pd.DataFrame.from_dict(jan_response)
liverpool_1
```
